```python
### Generate input files for wfreq with pure wstat eigenpotentials
import os

prefix = './'

for ecut in [50]:
    sample_in = prefix+'wfreq.in'
    sample_qsub = prefix+'wfreq.qsub'
    with open(sample_in, 'r') as inp, open(sample_qsub, 'r') as qsub:
        inp_lines = inp.readlines()
        qsub_lines = qsub.readlines()
        folder = prefix
        if not os.path.exists(folder):
            os.mkdir(folder)
        
        for n in range(100,1050,100):
            folder_ref = folder +'/std-'+str(n)
            if not os.path.exists(folder_ref):
                os.mkdir(folder_ref)
            ## Generate input files and qsub files
            file_inp = folder_ref+'/std-'+str(n)+'.in'
            outdir = "'./'"
            qe_prefix = "'si'"
            west_prefix = "'si_1000'"
            
            file_qsub = folder_ref+'/std-'+str(n)+'.qsub'
            job_name = 'std-'+str(n)
            
            with open(file_inp, 'w') as f, open(file_qsub, 'w') as g:
                for i in range(len(inp_lines)):
                    if 'qe_prefix' in inp_lines[i]:
                        f.write('qe_prefix           = '+qe_prefix+'\n')
                    elif 'west_prefix' in inp_lines[i]:
                        f.write('west_prefix         = '+west_prefix+'\n')
                    elif 'outdir' in inp_lines[i]:
                        f.write('outdir              = '+outdir+'\n')
                    elif 'n_pdep_eigen ' in inp_lines[i]:
                        f.write('n_pdep_eigen        = 1000\n')
                    elif 'n_pdep_eigen_to_use' in inp_lines[i]:
                        f.write('n_pdep_eigen_to_use  = '+str(n)+'\n')
                    elif 'trev_secant' in inp_lines[i]:
                        f.write('trev_secant          = '+str(ecut/13.6/1000)+'\n')
                    else:
                        f.write(inp_lines[i])
                for i in range(len(qsub_lines)):
                    if 'job-name' in qsub_lines[i]:
                        g.write('#SBATCH --job-name='+job_name+'\n')
                    elif 'mpirun' in qsub_lines[i]:
                        g.write('mpirun -n 160 /project/gagalli/yanghan/midway2/qe-6.1/bin/wfreq.x -nimage 16 < '
                                +'std-'+str(n)+'.in'+' >& '+'std-'+str(n)+'.out'+'\n')
                    else:
                        g.write(qsub_lines[i])
            

        
```
